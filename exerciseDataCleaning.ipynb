# Importing Module

```python
import pandas as pd
```

# Read CSV

```python
customer_data = pd.read_csv('Data/customerinfo.csv')
display(customer_data)
```

```python
transaction_data = pd.read_csv('Data/transaction.csv')
display(transaction_data)
```

# Find Duplicate

```python
culumns = ['first_name','last_name','gender','email','phone_number','address','city','country']
customer_duplicate_records = customer_data.duplicated(subset=culumns, keep=False)
duplicate_records = customer_data[customer_duplicate_records]
sorted_duplicate_records = duplicate_records.sort_values(by=['last_name','customer_id'])
sorted_duplicate_records
```

# Drop Duplicate

```python
customer_data.drop_duplicates(subset=culumns, keep='first', inplace=True)
customer_data.shape
```

# Find Null Sum

```python
missing_data = customer_data.isnull().sum()
missing_data
missing_data[missing_data > 0]
```

# Find NaN

```python
transaction_data[transaction_data.isna().any(axis=1)].shape
```

# Fill NaN

```python
transaction_data['total_amount'] = pd.to_numeric(transaction_data['total_amount'],errors='coerce')
transaction_data['unit_price'] = pd.to_numeric(transaction_data['unit_price'],errors='coerce')
transaction_data.fillna(
    value={
            'product_name' : transaction_data['product_name'].mode()[0],
            'quantity' : transaction_data['quantity'].mode()[0],
            'unit_price' : transaction_data['unit_price'].mean(),
            'total_amount' : transaction_data['total_amount'].mean(),
            'date' : transaction_data['date'].mode()[0]
           },
    inplace=True
)

transaction_data.shape
```

# Drop NaN

```python
transaction_data = transaction_data.dropna(subset=['total_amount','date'])
transaction_data.shape
```

# Merge

```python
merge_data = pd.merge(customer_data,transaction_data,on='customer_id')
sorted_data = merge_data.sort_values(by=['last_name'])
sorted_data
```

# Save Merge Data

```python
merge_data.to_csv('Data/finalldata.csv', index=False)
```
